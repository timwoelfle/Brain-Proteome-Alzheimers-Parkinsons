# Scatter Chart Examples

This notebook contains a number of examples of scatter-plot charts.

```python
import altair as alt
```

```python
# Uncomment/run this line to enable Altair in the classic notebook (not JupyterLab)
alt.renderers.enable('notebook')
```

```python
from vega_datasets import data
cars = data.cars()
cars.head()
```

## Dot chart

```python
alt.Chart(cars).mark_point().encode(
    x='Horsepower',
)
```

## Interactive Scatter chart

```python
alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon'
).interactive()
```

## Bubble chart

```python
alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    size='Acceleration'
)
```

## Scatter chart with color and/or shape

```python
alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    shape='Origin',
)
```

## Circle scatterchart

```python
alt.Chart(cars).mark_circle().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    tooltip='Origin'
)
```

## Square scatter chart

```python
alt.Chart(cars).mark_square().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
)
```

## Text scatter chart

```python
alt.Chart(cars).mark_text().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    text='Origin'
)
```

## Tick chart

```python
alt.Chart(cars).mark_tick().encode(
    x='Horsepower:Q',
    y='Cylinders:O',
)
```

## Multi-Panel Scatter

```python
chart = alt.Chart(cars).mark_point().encode(
    y='Horsepower:Q',
    color='Origin:N'
).interactive()

chart.encode(x='Acceleration:Q') | chart.encode(x='Displacement:Q')
```

## Interactive Repeated Scatter

```python
alt.Chart(cars).mark_point().encode(
    alt.X(alt.repeat('column'), type='quantitative'),
    alt.Y(alt.repeat('row'), type='quantitative'),
    color='Origin:N'
).properties(
    width=250,
    height=250
).repeat(
    row=['Horsepower', 'Miles_per_Gallon'],
    column=['Acceleration', 'Displacement']
).interactive()
```

## Multi-Panel Scatter with Linked Brushing

```python
brush = alt.selection_interval()

alt.Chart(cars).mark_point().encode(
    alt.X(alt.repeat('column'), type='quantitative'),
    alt.Y(alt.repeat('row'), type='quantitative'),
    color=alt.condition(brush, 'Origin:N', alt.value('gray'))
).add_selection(
    brush
).properties(
    width=250,
    height=250,
).repeat(
    row=['Horsepower', 'Miles_per_Gallon'],
    column=['Acceleration', 'Displacement']
)
```
